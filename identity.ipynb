```python
from analysis.pulse import analyze, cosine_similarity
from model import get_spikes
from analysis.musical import CONSONANCE_ORDER, calculate_intervals

def analyze_across_notes(root_note="C4"):
    intvl_mapping = calculate_intervals(root_note)
    spikes_list = [get_spikes(intvl_mapping[intvl]) for intvl in CONSONANCE_ORDER]
    analyze(spikes_list, cosine_similarity)
analyze_across_notes()
```
