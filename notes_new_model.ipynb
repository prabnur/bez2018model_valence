```python
from cache import SPIKES_DIR
import os
import numpy as np

newpath = os.path.join(SPIKES_DIR, "new", "piano")

C = np.load(os.path.join(newpath, "C4.npy"))
D = np.load(os.path.join(newpath, "D4.npy"))
```

```python
from scipy.io import loadmat

CFs = loadmat("cached/CFs.mat")["CFs"][0]
```

```python
from analysis.spike_tensor import generate_spike_tensor

single_tensor = generate_spike_tensor(C, duration=0.3)
mult_tensor = generate_spike_tensor(D, duration=0.3)
```

```python
from analysis.spectral import fourier_note

fourier_note("C4", shouldPlot=True)
```

```python

import numpy as np
from matplotlib.pyplot import scatter, axvline

cutoff = 2010
x_points = [263, 550, 800, 1100, 2000]

summed = np.sum(single_tensor, axis=(1, 2))
scatter(CFs[:cutoff], summed[:cutoff])
for x in x_points:
    axvline(x=x, color='r', linestyle=':')  # 'r' denotes red color
```

```python
summed = np.sum(mult_tensor, axis=(1, 2))
scatter(CFs[:cutoff], summed[:cutoff])
```
