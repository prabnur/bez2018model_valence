```python
from matplotlib import pyplot as plt
import numpy as np
from analysis.spatial import count_spikes_optimized

tone_spikes = get_tone_spikes(440)
spike_counts = count_spikes_optimized(tone_spikes)

def plot_counts(counts):
    counts_short = counts
    x = np.arange(len(counts_short))
    plt.plot(x, counts_short)
    plt.xlabel("Spike Train")
    plt.ylabel("Counts")
    plt.show()

plot_counts(spike_counts)
```

```python
tone_spikes = get_tone_spikes(2000)
spike_counts = count_spikes_optimized(tone_spikes)
plot_counts(spike_counts)
```
